```python
import requests
import datetime

app_id = '72a50186c43c49eea3bedb7e6a66c03f'
url = f"https://openexchangerates.org/api/latest.json?app_id={app_id}"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    base = data["base"]
    timestamp = data["timestamp"]
    rates = data["rates"]

    # Convert UNIX timestamp to readable format
    readable_time = datetime.datetime.fromtimestamp(timestamp)

    print(f"Base Currency: {base}")
    print(f"Timestamp: {timestamp} ({readable_time})")
    print("\nExchange Rates (USD to others):\n")

    for currency, rate in rates.items():
        print(f"USD to {currency}: {rate}")
else:
    print("Failed to fetch exchange rates:", response.text)
```
