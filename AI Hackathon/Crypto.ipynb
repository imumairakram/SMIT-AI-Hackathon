```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://coinmarketcap.com/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Get top 10 cryptocurrency rows
rows = soup.select("table tbody tr")[:10]

crypto_data = []

for row in rows:
    cols = row.find_all("td")
    if len(cols) < 9:
        continue

    rank = cols[1].text.strip()
    name = cols[2].find_all("p")[0].text.strip()
    symbol = cols[2].find_all("p")[1].text.strip()
    price = cols[3].text.strip()
    change_1h = cols[4].text.strip()
    change_24h = cols[5].text.strip()
    change_7d = cols[6].text.strip()

    # Extract second (full) Market Cap value
    raw_market_cap = cols[7].text.strip()
    market_cap_match = re.findall(r"\$\d{1,3}(?:,\d{3})*", raw_market_cap)
    market_cap = market_cap_match[-1] if market_cap_match else "N/A"

    # Extract only the first proper currency value from volume
    raw_volume = cols[8].text.strip()
    volume_match = re.search(r"\$\d{1,3}(?:,\d{3})*", raw_volume)
    volume_24h = volume_match.group(0) if volume_match else "N/A"

    crypto_data.append({
        "Rank": rank,
        "Name": name,
        "Symbol": symbol,
        "Price": price,
        "1h%": change_1h,
        "24h%": change_24h,
        "7d%": change_7d,
        "Market Cap": market_cap,
        "24h Volume": volume_24h
    })

df = pd.DataFrame(crypto_data)
df
```
